Select Victoria, download shp for selected records

```python
# Read the shapefile
shapefile_path = 'data/raw/georef-australia-state-suburb/georef-australia-state-suburb-millesime.shp'
gdf = gpd.read_file(shapefile_path)

# Display the first few records to understand its structure
gdf.head()
```

```python
import re

# Function to clean the strings by removing special characters and converting to lowercase
def clean_text(text):
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text).lower()  # Remove special characters
    return text.strip()

# Apply the function to clean both lga_name_lower and scc_name_lower columns
gdf['lga'] = gdf['lga_name'].apply(lambda x: clean_text(str(x)))
gdf['suburb'] = gdf['scc_name'].apply(lambda x: clean_text(str(x)))
lga_data = gdf[['suburb', 'lga']]
```

```python
lga_data.head()
```

```python
# Create a function to remove " vic" from suburb names
def remove_vic_suffix(suburb):
    if suburb.endswith(" vic"):
        return suburb.replace(" vic", "")
    return suburb

# Assuming the DataFrame is named 'lga_data' based on the screenshot
lga_data['suburb'] = lga_data['suburb'].apply(remove_vic_suffix)
```
