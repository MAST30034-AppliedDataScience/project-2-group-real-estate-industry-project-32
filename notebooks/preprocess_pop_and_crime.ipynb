## 2024-2027 Population Forecast

```python
data = merged_df

# Get a list of unique LGAs
lgas = data['lga'].unique()

# Check the range of years to ensure data consistency for forecasting
year_range = data['year'].min(), data['year'].max()

lgas, year_range
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

# Create a dictionary to store the forecasts
forecasts = {}

# Adjust the forecast retrieval process and rerun the forecast for 2024
for lga in lgas:
    # Filter the data for the current LGA
    lga_data = data[data['lga'] == lga].set_index('year')
    
    # Ensure no missing values and only population is used
    lga_population = lga_data['lga_population'].dropna()
    
    # Fit ARIMA model (Auto ARIMA can be done to select the best order, but here we use a simple approach)
    model = ARIMA(lga_population, order=(1, 1, 1))  # Using a (1,1,1) ARIMA model for simplicity
    model_fit = model.fit()
    
    # Forecast the population for 2024 (1 year ahead)
    forecast = model_fit.forecast(steps=1)
    
    # Store the forecast
    forecasts[lga] = forecast.iloc[0]  # Use iloc to correctly retrieve the forecast

# Convert forecasts to a DataFrame for easy viewing
forecast_df = pd.DataFrame(list(forecasts.items()), columns=['lga', 'forecast_2024'])
forecast_df
```

```python
# Add the forecast results for 2024 into the initial dataset
forecast_2024_df = pd.DataFrame({'lga': forecast_df['lga'], 'year': 2024, 'lga_population': forecast_df['forecast_2024']})

# Append the forecast results to the original dataset
pop_df = pd.concat([data, forecast_2024_df], ignore_index=True)
pop_df = pop_df.sort_values(by=['lga', 'year'])
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
import pandas as pd

# Create a dictionary to store the forecasts
forecasts = {}

# Adjust the forecast retrieval process and rerun the forecast for 2024 to 2027
for lga in lgas:
    # Filter the data for the current LGA
    lga_data = data[data['lga'] == lga].set_index('year')
    
    # Ensure no missing values and only population is used
    lga_population = lga_data['lga_population'].dropna()
    
    # Fit ARIMA model (Auto ARIMA can be done to select the best order, but here we use a simple approach)
    model = ARIMA(lga_population, order=(1, 1, 1))  # Using a (1,1,1) ARIMA model for simplicity
    model_fit = model.fit()
    
    # Forecast the population for 2024 to 2027 (4 years ahead)
    forecast = model_fit.forecast(steps=4)
    
    # Store the forecasts for 2024, 2025, 2026, and 2027
    forecasts[lga] = forecast

# Convert forecasts to a DataFrame for easy viewing
forecast_df = pd.DataFrame(forecasts).T
forecast_df.columns = ['forecast_2024', 'forecast_2025', 'forecast_2026', 'forecast_2027']

# Reset index and rename the first column to 'lga'
forecast_df = forecast_df.reset_index().rename(columns={'index': 'lga'})

forecast_df
```

## Forecast accuracy

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Initialize lists to store error metrics for each LGA
mae_list = []
mse_list = []
rmse_list = []

# Iterate through each LGA and compute error metrics
for lga in lgas:
    # Filter the data for the current LGA
    lga_data = data[data['lga'] == lga].set_index('year')
    
    # Ensure no missing values and only population is used
    lga_population = lga_data['lga_population'].dropna()
    
    # Fit ARIMA model
    model = ARIMA(lga_population, order=(1, 1, 1))
    model_fit = model.fit()
    
    # Perform in-sample forecast (fitted values)
    fitted_values = model_fit.fittedvalues
    
    # Compute errors
    mae = mean_absolute_error(lga_population[1:], fitted_values)
    mse = mean_squared_error(lga_population[1:], fitted_values)
    rmse = np.sqrt(mse)
    
    # Store the errors
    mae_list.append(mae)
    mse_list.append(mse)
    rmse_list.append(rmse)

# Create a DataFrame to display the accuracy metrics
accuracy_df = pd.DataFrame({
    'LGA': lgas,
    'MAE': mae_list,
    'MSE': mse_list,
    'RMSE': rmse_list
})
accuracy_df
```

```python
# Calculate overall accuracy metrics (mean of MAE, MSE, RMSE across all LGAs)
overall_mae = np.mean(mae_list)
overall_mse = np.mean(mse_list)
overall_rmse = np.mean(rmse_list)

overall_accuracy = {
    'Overall MAE': overall_mae,
    'Overall MSE': overall_mse,
    'Overall RMSE': overall_rmse
}

overall_accuracy
```

The overall MAE of 7,196.67 means that on average the model’s population predictions were off by around 7,197 people per LGA.

The overall RMSE of 22,059.42 indicates that the typical prediction error across all LGAs is around 22,059 people.

While some LGAs had relatively small errors, others had much larger deviations, which inflated the overall metrics.

### Forecast to 2027

```python
crime_data = df
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Prepare data for ARIMA forecasting
lgas = crime_data['Local Government Area'].unique()

# Create a dictionary to store the forecasts
forecasts = {}

# Adjust the forecast retrieval process and rerun the forecast for 2025 to 2027
for lga in lgas:
    # Filter the data for the current LGA
    lga_data = crime_data[crime_data['Local Government Area'] == lga].set_index('Year')
    
    # Ensure no missing values and only 'Rate per 100,000 population' is used
    lga_population = lga_data['Rate per 100,000 population'].dropna()
    
    # Fit ARIMA model (Auto ARIMA could be used, but we'll use a simple (1,1,1) model)
    model = ARIMA(lga_population, order=(1, 1, 1))
    model_fit = model.fit()
    
    # Forecast the population for 2025 to 2027 (3 years ahead)
    forecast = model_fit.forecast(steps=3)
    
    # Store the forecasts for 2025, 2026, and 2027
    forecasts[lga] = forecast

# Convert forecasts to a DataFrame for easy viewing
forecast_df = pd.DataFrame(forecasts).T
forecast_df.columns = ['forecast_2025', 'forecast_2026', 'forecast_2027']

# Reset index and rename the first column to 'lga'
forecast_df = forecast_df.reset_index().rename(columns={'index': 'lga'})
```

```python
forecast_df
```

```python
# Convert forecast data to a format similar to the original dataset
forecast_2025_df = pd.DataFrame({'Local Government Area': forecast_df['lga'], 'Year': 2025, 'Rate per 100,000 population': forecast_df['forecast_2025']})
forecast_2026_df = pd.DataFrame({'Local Government Area': forecast_df['lga'], 'Year': 2026, 'Rate per 100,000 population': forecast_df['forecast_2026']})
forecast_2027_df = pd.DataFrame({'Local Government Area': forecast_df['lga'], 'Year': 2027, 'Rate per 100,000 population': forecast_df['forecast_2027']})

# Append the forecasted data to the original dataset
combined_df = pd.concat([crime_data, forecast_2025_df, forecast_2026_df, forecast_2027_df], ignore_index=True)

# Sort the combined data by 'lga' and 'year'
combined_df = combined_df.sort_values(by=['Local Government Area', 'Year'])

# Display the combined dataframe
combined_df
```
