# Population

```python
file_path = 'data/raw/ABS_ERP_LGA2023_1.0.0_.LGA2023..A.csv'
school_quality_df = pd.read_csv(file_path)
school_quality_df.head()
```

```python
selected_columns = school_quality_df[['REGION: Region', 'TIME_PERIOD: Time Period', 'OBS_VALUE']]
selected_columns['REGION: Region'] = selected_columns['REGION: Region'].str.split(': ').str[1]
selected_columns['REGION: Region'] = selected_columns['REGION: Region'].str.lower()

# if lga name ends with "(vic)", remove that
selected_columns['REGION: Region'] = selected_columns['REGION: Region'].str.replace(' \(vic.\)', '', regex=True)

# Rename the columns
selected_columns.rename(columns={
    "REGION: Region": "lga",
    "TIME_PERIOD: Time Period": "year",
    "OBS_VALUE": "lga_population"
}, inplace=True)

selected_columns
```

```python
selected_columns.to_csv('data/curated/lga_pop.csv')
```

```python
lga_pop = selected_columns
```

# Merge into VIC LGA list (check for bugs)

```python
# Merge using a left join to ensure all rows from lga_check are preserved
merged_df = pd.merge(lga_check, lga_pop, left_on='lga', right_on='lga', how='left')
merged_df
```

```python
merged_df = merged_df.sort_values(by=['lga', 'year'])
merged_df
```

```python
#merged_df.to_csv('data/curated/lga_pop_bugcheck.csv')
```

```python
merged_df = merged_df.dropna() # remove rows with empty population (national parks etc.)
merged_df = merged_df.reset_index(drop=True) # remove index
```

```python
merged_df
```

```python
merged_df.to_csv('data/curated/lga_pop.csv', index=False)
```

# Crime

```python
# Load the Excel file to inspect the available sheets
file_path = 'data/raw/Data_Tables_LGA_Recorded_Offences_Year_Ending_June_2024.xlsx'
xls = pd.ExcelFile(file_path)

# List all sheet names to identify "Table 02"
sheet_names = xls.sheet_names
sheet_names
```

```python
df_raw = pd.read_excel(file_path, sheet_name='Table 01')
df_raw
```

```python
nan_rate_rows = df_raw[df_raw['Rate per 100,000 population'].isna()]
nan_rate_rows.head(10)
```

```python
df = df_raw[['Local Government Area', 'Year', 'Rate per 100,000 population']]
df['Local Government Area'] = df['Local Government Area'].str.strip()  # Remove any leading/trailing whitespace
df = df[~df['Local Government Area'].isin(['Total', 'Justice Institutions and Immigration Facilities', 'Unincorporated Vic'])]
df.head()
```

```python
df['Local Government Area'] = df['Local Government Area'].str.lower()  # Convert to lowercase
df = df.sort_values(by=['Local Government Area', 'Year'])
df
```

```python
df.to_csv('data/curated/crime_by_lga.csv')
```
