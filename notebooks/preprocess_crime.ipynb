```python
import pandas as pd
```

https://www.crimestatistics.vic.gov.au/crime-statistics/latest-victorian-crime-data/download-data

```python
# Load the Excel file to inspect the available sheets
file_path = 'data/raw/Data_Tables_LGA_Recorded_Offences_Year_Ending_June_2024.xlsx'
xls = pd.ExcelFile(file_path)

# List all sheet names to identify "Table 02"
sheet_names = xls.sheet_names
sheet_names
```

```python
df_raw = pd.read_excel(file_path, sheet_name='Table 01')
df_raw
```

```python
df_raw.to_csv('data/curated/crime_v1.csv')
```

```python
nan_rate_rows = df_raw[df_raw['Rate per 100,000 population'].isna()]
nan_rate_rows.head(10)
```

```python
df = df_raw[['Local Government Area', 'Year', 'Rate per 100,000 population']]
df['Local Government Area'] = df['Local Government Area'].str.strip()  # Remove any leading/trailing whitespace
df = df[~df['Local Government Area'].isin(['Total', 'Justice Institutions and Immigration Facilities', 'Unincorporated Vic'])]
df.head()
```

```python
df['Local Government Area'] = df['Local Government Area'].str.lower()  # Convert to lowercase
df = df.sort_values(by=['Local Government Area', 'Year'])
df
```

```python
df.to_csv('data/curated/crime_by_lga.csv')
```
